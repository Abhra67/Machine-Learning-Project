```python
import pandas as pd

# Dataset load koro
df = pd.read_csv("/content/Spotify Million Song Dataset_exported.csv")  # Path change koro jodi lagbe

# First few rows check koro
df.head()
```

```python
# Duplicate remove koro
df = df.drop_duplicates(subset=['artist', 'song'])

# Jodi lyrics missing thake, remove koro
df = df.dropna(subset=['text'])

# Index reset kore dao
df = df.reset_index(drop=True)

# Data size check koro
df.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# TF-IDF model create
vectorizer = TfidfVectorizer(stop_words='english')

# Lyrics ke TF-IDF vector e convert
tfidf_matrix = vectorizer.fit_transform(df['text'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def find_similar_song(input_text):
    # Input text ke TF-IDF vector e convert
    input_vec = vectorizer.transform([input_text])

    # Sob song er sathe similarity check
    similarities = cosine_similarity(input_vec, tfidf_matrix).flatten()

    # Best match er index ber koro
    best_match_index = similarities.argmax()

    # Match paoa song ar artist return koro
    return df.loc[best_match_index, ['song', 'artist']].to_dict()
```

```python
sample_input = "Take it easy with me, please"  # Test text
result = find_similar_song(sample_input)
print(result)  # Output dekho
```
